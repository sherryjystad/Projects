### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
# Find the amount of Players
unique_players_df = purchase_data.groupby('SN').nunique()
total_count = len(unique_players_df)
total_breakdown = pd.DataFrame({"Total Players": [total_count]
                                })
total_breakdown
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data['Item ID'].nunique()
average_price = purchase_data['Price'].mean()
total_purchases = len(purchase_data)
total_revenue = purchase_data['Price'].sum()
summary_breakdown = pd.DataFrame({"Number of Unique Items": [unique_items],
                                   "Average Price": [average_price],
                                   "Number of Purchases": [total_purchases],
                                   "Total Revenue": [total_revenue]
                                 })
summary_breakdown["Average Price"] = summary_breakdown["Average Price"].map(
    "${0:,.2f}".format)
summary_breakdown["Total Revenue"] = summary_breakdown["Total Revenue"].map(
    "${0:,.2f}".format)
summary_breakdown
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_dgs_df = purchase_data[["SN", "Gender", "Price"]] 
gender_players = gender_dgs_df.groupby('Gender')['SN'].nunique()
gender_percent = round((gender_players/total_count) *100,2)

gender_summary = pd.DataFrame({"Total Count": gender_players,
                                "Percentage of Players": gender_percent
                                })
gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map(
    "{0:,.2f}%".format)
gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase Count for gender
gender_purchase_count = purchase_data.groupby('Gender')['Purchase ID'].count()

# Average Purchase Price for gender
gender_purchase_avg = round(gender_dgs_df.groupby("Gender")["Price"].mean(),2)

# Total Purchase value for each gender
gender_purchase_value = round(gender_dgs_df.groupby('Gender')['Price'].sum(),2)

# Avg Total Purchase per Person
avg_purchase_per_person = round(gender_purchase_value/gender_players,2)

# Summary Data Frame
purchase_summary = pd.DataFrame({"Purchase Count": gender_purchase_count,
                                "Average Purchase Price": gender_purchase_avg,
                                 "Total Purchase Value": gender_purchase_value,
                                 "Average Total Purchase per Person": avg_purchase_per_person
                                })

purchase_summary["Average Purchase Price"] = purchase_summary["Average Purchase Price"].map(
    "${0:,.2f}".format)
purchase_summary["Total Purchase Value"] = purchase_summary["Total Purchase Value"].map(
    "${0:,.2f}".format)
purchase_summary["Average Total Purchase per Person"] = purchase_summary["Average Total Purchase per Person"].map(
    "${0:,.2f}".format)
purchase_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins in which to place values based Heroes of Pymoli
bins = [0, 9, 14, 19, 24, 29, 34, 39, 120]

# Create labels for these bins
group_labels = ["< 10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]

# Slice the data and place it into bins
pd.cut(purchase_data["Age"], bins, labels=group_labels).head()

# Place the data series into a new column inside of the DataFrame
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

# Create new data frame for Age Demographics
age_range = purchase_data.groupby("Age Range")["SN"].nunique()
age_summary = pd.DataFrame({"Age Range": age_range,
                            "Percentage of Players": round((age_range/total_count)*100,2)
                            })
age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase Count for Age Range
age_purchase_count = purchase_data.groupby("Age Range")["Purchase ID"].count()

# Age Count for Age Range and Name using nunique
age_count_unique = purchase_data.groupby("Age Range")["SN"].nunique()

# Age Average Purchase Price
age_average_price = round(purchase_data.groupby("Age Range")["Price"].mean(),2)

# Age Total Purchase Value
age_purchase_value = round(purchase_data.groupby("Age Range")["Price"].sum(),2)

# Age Average Total Purchase per Person
age_avg_total = round(age_purchase_value/age_count_unique,2)

# New dataframe for Age Purchasing Analysis
age_purchase_summary = pd.DataFrame({"Purchase Count": age_purchase_count,
                                "Average Purchase Price": age_average_price,
                                "Total Purchase Value": age_purchase_value,
                                 "Avg Total Purchase per Person": age_avg_total
                                })

age_purchase_summary["Average Purchase Price"] = age_purchase_summary["Average Purchase Price"].map(
    "${0:,.2f}".format)
age_purchase_summary["Total Purchase Value"] = age_purchase_summary["Total Purchase Value"].map(
    "${0:,.2f}".format)
age_purchase_summary["Avg Total Purchase per Person"] = age_purchase_summary["Avg Total Purchase per Person"].map(
    "${0:,.2f}".format)
age_purchase_summary
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create data frame and sort
popular_items_pd = pd.DataFrame(purchase_data, columns=["Item ID", "Item Name", "Item Price"])

#Calculations
popular_items = purchase_data.groupby("Item Name")['Item ID'].count() 
#print (popular_items)
popular_item_price = round(purchase_data.groupby("Item Name")['Price'].nunique(),2)
#print (popular_item_price)
popular_item_value = round(purchase_data.groupby("Item Name")["Price"].sum(),2)
#print (popular_item_value)

# New dataframe for Popular Items
popular_item_summary = pd.DataFrame({"Purchase Count": popular_items,
                                     "Item Price": popular_item_price,
                                     "Total Purchase Value": popular_item_value
                                  })
# Merge Frames
pd.merge(popular_items_pd, popular_item_summary , on=("Item Price"), how="right")
```
